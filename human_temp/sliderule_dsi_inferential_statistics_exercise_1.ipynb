Import the necessary libraries and examine the data frame structure and look at some sample data.

```python
#Import all the modules that are going to be used
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from  datetime import datetime
df = pd.read_csv('D:/SpringBoard/EDAProjects/1509032434_human_temp/human_temp/data/human_body_temperature.csv')
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
# Compute number of data points: n_data
n_data = len(df)
# Number of bins is the square root of number of data points: n_bins
n_bins = np.sqrt(n_data)
# Convert number of bins to integer: n_bins
n_bins = int(n_bins)

sns.set()
_ = plt.hist(df['temperature'],n_bins) 

_ = plt.ylabel('Count ')
_ = plt.xlabel ('Temperature')
plt.show() 
```

Plot of the Histogram with the standard bin size - the distribution is somewhat like a normal distribution ( one symmetrical peak).Proceed to plot the ECDF to see what the data shows

```python
sns.set()
x,y = ecdf(df['temperature'])

_ = plt.plot (x, y, marker = '.', linestyle = 'none')

# Make the margins nice
plt.margins(0.02)

# Label the axes
_ = plt.xlabel('Temperature of humans')
_ = plt.ylabel('ECDF')

# Display the plot
plt.show()

```

```python
# Plot the ECDFs for both the observed data and the sampled data that was obtained by the computed mean & std dev

mean = np.mean(df['temperature']) 
std = np.std(df['temperature']) 
samples = np.random.normal(mean,std,size=10000 )
x,y= ecdf(df['temperature'])
x_t, y_t = ecdf(samples)

_ = plt.plot (x, y, marker = '.', linestyle = 'none')
_ = plt.plot (x_t, y_t, marker = '.', linestyle = 'none')

# Make the margins nice
plt.margins(0.02)

# Label the axes
_ = plt.xlabel('Temperature of humans')
_ = plt.ylabel('ECDF')

# Display the plot
plt.show()
```
